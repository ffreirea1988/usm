**Gráficos**

```python
# 1 solo Histograma
# Con fig, ax
fig, ax=plt.subplots()
ax.hist(df['Weekly_Sales'])
ax.set_title('Histograma ventas semanales')
plt.show()
```

```python
#Histograma
# Con la función hist()
df.hist(figsize=(10,5), layout=(2,4))
plt.tight_layout()
plt.show()
```

```python
#Relación entre 2 variables: Weekly_Sales vs Fuel_Price
# Con fig, ax
fig, ax=plt.subplots()
ax.scatter(x=df['Fuel_Price'], y=df['Weekly_Sales'])
plt.show()
```

```python
#Relación entre 2 variables: Weekly_Sales vs Fuel_Price
#Con la función scatter()
plt.scatter(df['Fuel_Price'], df['Weekly_Sales'], s=1)
plt.title('Weekly Sales vs Fuel Price')
plt.xlabel('Fuel Price')
plt.ylabel('Weekly Sales')
plt.show()
```

```python
df.columns
```

```python
#Relación entre 2 variables: Weekly_Sales vs Fuel_Price
plt.scatter(df['Temperature'], df['Weekly_Sales'], s=1, color='green')
plt.show()
```

```python
#1 solo Boxplot
# Con fig, ax
fig, ax=plt.subplots()
ax.boxplot(df['Weekly_Sales'])
ax.set_title('Diagrama de caja y bigotes ventas semanales')
plt.show()
```

```python
#Boxplot
# Con la función boxplot()
plt.figure(figsize=(10,4))
df.boxplot()
plt.show()
```

```python
#1 solo Boxplot de otra variable
#Para revisar la dispersión de la temperatura
# Con fig, ax
fig, ax=plt.subplots()
ax.boxplot(df['Temperature'])
ax.set_title('Diagrama de caja y bigotes Temperatura')
plt.show()
```

```python
#Con la función plot()
df.plot(kind='box', subplots=True, layout=(3,4), figsize=(12,6))
plt.show()
```
